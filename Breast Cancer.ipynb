```python
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
```

```python
dataset.keys()
```

```python
dataset.target_names
```

```python
dataset.feature_names
```

```python
#df= pd.DataFrame (dataset.data,columns=dataset.feature_names)
#df
```

```python
dataset.target
```

```python
x=dataset.data
y=dataset.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test= train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_obj=DecisionTreeClassifier()
tree_obj
```

```python
tree_obj.fit(x_train,y_train)
```

```python
y_pred =  tree_obj.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# Pre Prunning

```python
parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']
    }
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_obj =GridSearchCV(tree_obj,param_grid=parameter,cv=5,scoring="accuracy")
grid_obj
```

```python
import warnings
warnings.filterwarnings('ignore')
grid_obj.fit(x_train,y_train)
```

```python
grid_obj.best_params_
```

```python
pre_obj=DecisionTreeClassifier(criterion="gini",max_depth=3,max_features="auto",splitter="best")
```

```python
pre_obj.fit(x_train,y_train)
```

```python
y_pred2=pre_obj.predict(x_test)
y_pred2
```

```python
y_test
```

```python
print(confusion_matrix(y_test,y_pred2))
print(accuracy_score(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
```

We can clearly see accuracy has been increased to 94%
